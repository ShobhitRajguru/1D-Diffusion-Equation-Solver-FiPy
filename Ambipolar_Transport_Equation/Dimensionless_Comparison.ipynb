```python
# ============================================================
# PART 1 : SOLVING THE DIMENSIONAL STEADY-STATE EQUATION
# ============================================================

import sympy as sp
from IPython.display import display, Math

# ------------------------------------------------------------
# Step 1 : Define Symbols
# ------------------------------------------------------------

x = sp.symbols('x', real=True)

Dp = sp.Symbol('D_p', positive=True)
mu_p = sp.Symbol('mu_p', positive=True)
E = sp.Symbol('E', real=True)

# Unknown function
p = sp.Function('p')

# ------------------------------------------------------------
# Step 2 : Define the Governing Equation
# Dp*d²p/dx² + μE*dp/dx = 0
# ------------------------------------------------------------

eq = sp.Eq(
    Dp*sp.diff(p(x), x, 2)
    +
    mu_p*E*sp.diff(p(x), x),
    0
)

print("Original Governing Equation:")
display(eq)

# ------------------------------------------------------------
# Step 3 : Solve the Governing Equation
# ------------------------------------------------------------

solution = sp.dsolve(eq)

print("\nAnalytical Solution:")
display(solution)
```

```python
# ============================================================
# PART 2 : SOLVING THE DIMENSIONLESS STEADY-STATE EQUATION
# ============================================================

import sympy as sp
from IPython.display import display

# ------------------------------------------------------------
# Step 1 : Define Symbols
# ------------------------------------------------------------

x_tilde = sp.symbols('x_tilde', real=True)

Dp = sp.Symbol('D_p', positive=True)
mu_p = sp.Symbol('mu_p', positive=True)
E = sp.Symbol('E', real=True)
L = sp.Symbol('L', positive=True)

# Unknown function
p_tilde = sp.Function('p_tilde')

# ------------------------------------------------------------
# Step 2 : Define the Dimensionless Governing Equation
#
# d²p~/dx~² + (μ_p E L / D_p) dp~/dx~ = 0
# ------------------------------------------------------------

eq_tilde = sp.Eq(
    sp.diff(p_tilde(x_tilde), x_tilde, 2)
    +
    (mu_p*E*L/Dp)*sp.diff(p_tilde(x_tilde), x_tilde),
    0
)

print("Dimensionless Governing Equation:")
display(eq_tilde)

# ------------------------------------------------------------
# Step 3 : Solve the Dimensionless Governing Equation
# ------------------------------------------------------------

solution_tilde = sp.dsolve(eq_tilde)

print("\nDimensionless Analytical Solution:")
display(solution_tilde)
```

```python
# ============================================================
# PART 3 : COMPARISON OF COMPLETE ANALYTICAL SOLUTIONS
# ============================================================

import sympy as sp
from IPython.display import display

# ------------------------------------------------------------
# Step 1 : Extract the RHS of both solutions
# ------------------------------------------------------------

dimensional_solution = solution.rhs
dimensionless_solution = solution_tilde.rhs

print("Dimensional Analytical Solution:")
display(dimensional_solution)

print("\nDimensionless Analytical Solution:")
display(dimensionless_solution)

# ------------------------------------------------------------
# Step 2 : Substitute x_tilde = x/L
# ------------------------------------------------------------

dimensionless_solution = dimensionless_solution.subs(x_tilde, x/L)

dimensionless_solution = sp.simplify(dimensionless_solution)

print("\nDimensionless Solution after substitution:")
display(dimensionless_solution)

# ------------------------------------------------------------
# Step 3 : Compare Both Solutions
# ------------------------------------------------------------

print("\nComparison:")

display(sp.Eq(dimensionless_solution, dimensional_solution))

print("\nDifference:")

difference = sp.simplify(dimensionless_solution - dimensional_solution)

display(difference)

# ------------------------------------------------------------
# Step 4 : Final Verification
# ------------------------------------------------------------

if difference == 0:
    print("\n✅ VERIFIED")
    print("The dimensional and dimensionless analytical solutions are identical.")
else:
    print("\n❌ Solutions are NOT identical.")
```
